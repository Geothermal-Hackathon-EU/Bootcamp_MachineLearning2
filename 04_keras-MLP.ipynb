```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
 
```

```python
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
```

```python
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))#inputense genera una rete fully connected con 12 = num di neuroni 
# input_dim= numero di dati di input --- se sono sul secondo layer Ã¨ automatico - oppure input layer 
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

Automatic Validation set

```python
# Fit the model
model.fit(X, Y, validation_split=0.33 ,epochs=150, batch_size=10)
```

manual train/test split

```python
#split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

```python
#fitting the model
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

## Saving & loading models 

```python
from keras.models import model_from_json
import numpy
import os
```

```python
# serialize model to JSON
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
print("Saved model to disk")
```

```python
#load from disk
json_file = open("model.json", "r")
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("model.h5")
print("Loaded model from disk")
```
